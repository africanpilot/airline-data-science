## Data Transformation and Data Analysis Ver_03

```python
#8 Data Transformation Ver_03

# merge and map data
display(md('## **Data Transformation and Data Analysis Ver_03**'))
display(md('###### **Merge ac_core, ac_region_observer, and ac_month as animal_crakers_merged**'))
animal_crakers_merged = ac_core.merge(ac_region_observer,on='ID').merge(ac_month,on='ID')
animal_crakers_merged_first = animal_crakers_merged.copy()
animal_crakers_merged_v3 = animal_crakers_merged.copy()
display(animal_crakers_merged.head())

display(md('###### **Map Species to "Type", "Type_x", and "Type_y" in animal_crakers_merged**'))
data_trans_info.df_map(df=animal_crakers_merged_v3,df_col='Type',df_map=ac_species_code,col_id='Species_Code',col_map='Species')
data_trans_info.df_map(df=animal_crakers_merged_v3,df_col='Type_x',df_map=ac_species_code,col_id='Species_Code',col_map='Species')
data_trans_info.df_map(df=animal_crakers_merged_v3,df_col='Type_y',df_map=ac_species_code,col_id='Species_Code',col_map='Species')
display(animal_crakers_merged_v3.head())

# Uppcase columns
display(md('**Verify Before and After results for Upper case columns**'))
animal_crakers_merged_v3 = animal_crakers_merged_v3.applymap(lambda s:s.upper() if type(s) == str else s)

# Before Results for Upper case columns
df_before_after_name = ["Before","After"]
df_before_after_df = [animal_crakers_merged,animal_crakers_merged_v3]
data_anomaly_info.df_unique(df_list=df_before_after_df,df_names=df_before_after_name)

#8.b Rename 'Type_x' to species and drop 'Type' and 'Type_y'
display(md('**Verify Before and After results for Rename and drop columns**'))
before_rename = animal_crakers_merged_v3.copy()
animal_crakers_merged_v3.rename(columns={'Type_x': 'Species'}, inplace=True)
animal_crakers_merged_v3 = animal_crakers_merged_v3.drop(columns=['Type_y', 'Type'])
df_before_after_name = ["Before","After"]
df_before_after_df = [before_rename,animal_crakers_merged_v3]
data_trans_info.df_html_head(df_list=df_before_after_df,df_names=df_before_after_name)

#8.b Convert Lbs to Kg
display(md('**Verify Before and After results for Weight_lb to Weight_kg conversion**'))
animal_crakers_merged_v3 = data_trans_info.lbs_to_kg_conversion(df=animal_crakers_merged_v3,kg_col='Weight_kg', lbs_col='Weight_lb')
move_col = animal_crakers_merged_v3.columns.tolist()
move_col.insert(3, move_col.pop(move_col.index('Weight_kg')))
animal_crakers_merged_v3 = animal_crakers_merged_v3.reindex(columns= move_col)
display(animal_crakers_merged_v3.head())

#8.c Identify empty cells, replace with mode value by species group
display(md('**Verify Before and After results for filling missing values with groupped mode**'))

# group species and find mode for each
animal_crakers_merged_v3_mode = animal_crakers_merged_v3[['Species','Weight_lb','Weight_kg','Intelligence']]
species_grouped = data_trans_info.grouped_mode(df=animal_crakers_merged_v3_mode,group_col='Species')
display(species_grouped)

mergered_before_missing_nulls = data_anomaly_info.df_nulls(df=animal_crakers_merged_v3)
mergered_after_missing_nulls = data_anomaly_info.df_nulls(df=animal_crakers_merged_v3)

# Map grouped mode values to missing values for weight_lb, weight_kg, and intelligence
map_col = ['Weight_lb','Weight_kg','Intelligence']
animal_crakers_merged_v3 = data_trans_info.grouped_mapping(df=animal_crakers_merged_v3,df_col=map_col,df_map=species_grouped,df_id='Species')
before_robust_scaler = animal_crakers_merged_v3.copy()
mergered_after_missing_nulls = data_trans_info.grouped_mapping(df=mergered_after_missing_nulls,df_col=map_col,df_map=species_grouped,df_id='Species')

# verify before and after results for filling weight and intelligence with mode per species
animal_crakers_merged_v3_nulls = data_anomaly_info.df_nulls(df=animal_crakers_merged_v3)
df_before_after_name = ["Before","After","Remain"]
df_before_after_df = [mergered_before_missing_nulls,mergered_after_missing_nulls,animal_crakers_merged_v3_nulls]
data_trans_info.df_html(df_list=df_before_after_df,df_names=df_before_after_name)

#Verify before and after robust scaler
display(md('### **Verify Before and After results for Robust Scaler and then Standard deviation**'))

# Use RobustScaler to set a 0-1 scale
scaler = preprocessing.RobustScaler()
animal_crakers_merged_v3[['Weight_lb','Weight_kg', 'Intelligence']] = scaler.fit_transform(animal_crakers_merged_v3[['Weight_lb','Weight_kg', 'Intelligence']])

before_robust_scaler_html = before_robust_scaler[['Weight_lb','Weight_kg', 'Intelligence']].head()
animal_crakers_merged_v3_robust_html = animal_crakers_merged_v3[['Weight_lb','Weight_kg', 'Intelligence']].head()

df_before_after_name = ["Before","After"]
df_before_after_df = [before_robust_scaler_html,animal_crakers_merged_v3_robust_html]
data_trans_info.df_html(df_list=df_before_after_df,df_names=df_before_after_name)

title_plt1 = 'Intelligence Density Plot: Before Robust Scaler'
title_plt2 = 'Intelligence Density Plot: After Robust Scaler'
data_trans_info.kde_subplot(df1=before_robust_scaler,df2=animal_crakers_merged_v3,hue='Species',df_col='Intelligence'
                            ,title1=title_plt1,title2=title_plt2)

title_plt1 = 'Weight_kg Density Plot: Before Robust Scaler'
title_plt2 = 'Weight_kg Density Plot: After Robust Scaler'
data_trans_info.kde_subplot(df1=before_robust_scaler,df2=animal_crakers_merged_v3,hue='Species',df_col='Weight_kg'
                            ,title1=title_plt1,title2=title_plt2)

#8.e Filter outliers more than 3 Standard Deviation
display(md('**Verify Before and After results for Standard Deviations**'))
before_std_dev = animal_crakers_merged_v3.copy()
before_removed_df = animal_crakers_merged_v3.copy()
df_col = ['Intelligence','Weight_kg','Weight_lb']

std_removed_df = data_trans_info.standard_deviation_removed(df=before_removed_df,group='Species',column=df_col[0],stand_dev=3)
std_removed_df = data_trans_info.standard_deviation_removed(df=before_removed_df,group='Species',column=df_col[1],stand_dev=3)
std_removed_df = data_trans_info.standard_deviation_removed(df=before_removed_df,group='Species',column=df_col[2],stand_dev=3)
std_removed_df = std_removed_df[std_removed_df['Species'].notnull()]

animal_crakers_merged_v3 = data_trans_info.standard_deviation_grouped(df=animal_crakers_merged_v3,group='Species',column=df_col,stand_dev=3)
animal_crakers_merged_v3 = animal_crakers_merged_v3[np.isfinite(animal_crakers_merged_v3['ID'])]
finial_df = data_anomaly_info.df_nulls(df=animal_crakers_merged_v3)

df_before_after_name = ["Removed","Remaining"]
df_before_after_df = [std_removed_df,finial_df]
data_trans_info.df_html(df_list=df_before_after_df,df_names=df_before_after_name)


# Verify before and after for standard devation
title_plt1 = 'Intelligence Density Plot: Before Standard Devation'
title_plt2 = 'Intelligence Density Plot: After Standard Devation'
data_trans_info.kde_subplot(df1=before_std_dev,df2=animal_crakers_merged_v3,hue='Species',df_col='Intelligence'
                            ,title1=title_plt1,title2=title_plt2)

title_plt1 = 'Weight_kg Density Plot: Before Standard Devation'
title_plt2 = 'Weight_kg Density Plot: After Standard Devation'
data_trans_info.kde_subplot(df1=before_std_dev,df2=animal_crakers_merged_v3,hue='Species',df_col='Weight_kg'
                            ,title1=title_plt1,title2=title_plt2)

#4.f Scatter Plot: Before  and after for weight and intelligence for each species
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False, sharex = False, figsize=(15,5))
sns.scatterplot(x="Weight_lb", y="Intelligence", hue="Species", data=before_robust_scaler, ax=ax1).set_title('Before Robust Scaler: Weight and Intelligence Scatter Plot');
sns.scatterplot(x="Weight_lb", y="Intelligence", hue="Species", data=before_std_dev, ax=ax2).set_title('After Robust Scaler: Weight and Intelligence Scatter Plot');

#4.f Scatter Plot: Before  and after for weight and intelligence for each species
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=False, sharex = False, figsize=(15,5))
sns.scatterplot(x="Weight_lb", y="Intelligence", hue="Species", data=before_std_dev, ax=ax1).set_title('Before Standard Deviation: Weight and Intelligence Scatter Plot');
sns.scatterplot(x="Weight_lb", y="Intelligence", hue="Species", data=animal_crakers_merged_v3, ax=ax2).set_title('After Standard Deviation: Weight and Intelligence Scatter Plot');

#Data useage anaysis
#9 Percentage of kept data
display(md('**Data Usage analysis Ver_03**'))
df_before_after_name = ["animal_crakers_merged_v3"]
df_inital = [animal_crakers_merged_first]
df_final = [animal_crakers_merged_v3]
data_anomaly_info.df_row_count(df_int=df_inital,df_fin=df_final,df_names=df_before_after_name)
```
